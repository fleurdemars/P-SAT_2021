```r
# 시도 이름
do = c('부산광역시', '충청북도', '충청남도', '대구광역시', '대전광역시', '강원도',
      '광주광역시', '경상북도', '경기도', '경상남도', '인천광역시', '제주특별자치도', 
       '전라북도', '전라남도', '세종특별자치시', '서울특별시', '울산광역시')
```

```r
data0 = data.frame()

for (i in 1:length(list_grdp)){
    data = cbind(list_grdp[[i]], income[income['시도별']==do[i],])[-1,-c(3,4)]
    rownames(data) = NULL
    data = left_join(data, list_popl[[i]], key = 'sigungu')
    data = left_join(data, list_fam[[i]], key = 'sigungu')
    data$chong = data$chong %>% as.numeric()
    data$grdp = data$grdp %>% as.numeric()
    data$grdp = data$grdp / data$chong
    
    data0 = rbind(data0, data)
}
```
